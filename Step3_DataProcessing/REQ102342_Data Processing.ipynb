```python
# read the data from the "2022_2023-quarterly-surgical_wait_times_2018-Onwards.csv" file
# and calculate the coefficient of variation based on the "COMPLETED_50TH_PERCENTILE" column

import pandas as pd

# Read the data from the CSV file
df = pd.read_csv("2022_2023-quarterly-surgical_wait_times_2018-Onwards.csv")

# Group the data by "procedure_group"
grouped_df = df.groupby("PROCEDURE_GROUP")

# Calculate the coefficient of variation for each "procedure_group"
coefficient_of_variation = grouped_df["COMPLETED_50TH_PERCENTILE"].std() / grouped_df["COMPLETED_50TH_PERCENTILE"].mean()

# Find the procedure_group with the greatest fluctuation
greatest_fluctuation_procedure_group = coefficient_of_variation.idxmax()
greatest_fluctuation_value = coefficient_of_variation.max()

print(f"The procedure_group with the greatest fluctuation based on COMPLETED_50TH_PERCENTILE is '{greatest_fluctuation_procedure_group}' with a coefficient of variation of {greatest_fluctuation_value:.2f}.")
```

```python
# Identify the top three procedure_groups with the greatest/least fluctuation
# based on the COMPLETED_50TH_PERCENTILE column

import pandas as pd

# Read the data from the CSV file
df = pd.read_csv("2022_2023-quarterly-surgical_wait_times_2018-Onwards.csv")

# Group the data by "procedure_group"
grouped_df = df.groupby("PROCEDURE_GROUP")

# Calculate the coefficient of variation for each "procedure_group"
coefficient_of_variation = grouped_df["COMPLETED_50TH_PERCENTILE"].std() / grouped_df["COMPLETED_50TH_PERCENTILE"].mean()

# Sort the coefficient of variation in descending order
sorted_coeff_variation = coefficient_of_variation.sort_values(ascending=False)

# Get the top three procedure_groups with the greatest fluctuation
top_three_procedure_groups = sorted_coeff_variation.index[:3]
top_three_values = sorted_coeff_variation.values[:3]

print("Top Three Procedure Groups with the Greatest Fluctuation based on COMPLETED_50TH_PERCENTILE:")
for i, group in enumerate(top_three_procedure_groups):
    print(f"{i+1}. Procedure Group: {group}, Coefficient of Variation: {top_three_values[i]:.2f}")
```
